# Importamos Librerias

```python
import numpy as np
import random
import copy
import matplotlib.pyplot as plt
```

# Juego del Caos

```python
#Definimos los puntos
a=[(1,1),(2,2),(3,1)]

#Creamos la figura
fig = plt.figure(figsize=(8, 8))

#Seleccionamos punto inicial aleatorio y lo graficamos
r=random.randint(0,len(a)-1)
aux=copy.deepcopy(a[r])
plt.scatter(aux[0],aux[1])
print("Original",aux,"con r",r)

#Juego del caos
for i in range(1000):
  r=random.randint(0,len(a)-1)
  aux=[(aux[0]+a[r][0])/2,(aux[1]+a[r][1])/2]
  plt.scatter(aux[0],aux[1])
#Mostramos resultados
plt.show()
```

# Importamos Librerias

```python
#Importamos librerias
import sympy
import copy
import matplotlib.pyplot as plt
import random
import PIL
from PIL import Image, ImageDraw
```

#Graficacion

```python
#Establecemos x como variable
x=sympy.symbols("x")


#f=(8+4j)*x**2+x+1
#f=x**5+x**4+x**3+x+1
f=x**2+(-.79+.15j)
#f=x**2+(-.12-.77j)
#f=x**2+(-1.476+0j)
#f=x**2+(-.162+1.04j)
#f=x**2+(.28+.008j)

#Determina el orden de la ecuacion
try:
  t = sympy.Poly(f,x)
  if(len(t.all_coeffs())==3):
    ord2=True
  else:
    ord2=False
except:
  ord2=False
#La hacemos mas eficiente
fef = sympy.lambdify(x, f, modules="numpy")


# create an image
ancho,alto=200,200
out = Image.new(mode = "RGB", size = (ancho, alto), color = (255, 255, 255))
#Condicion de paro
paro=random.randint(1,10)
if ord2:
  paro=2

#Graficamos el conjunto
for i in range(ancho):
  for j in range(alto):
    a=(i/alto)*2-1
    b=(j/ancho)*2-1
    res=complex(a,b)
    for k in range(1000):
      res=fef(res)
      #Cambio de color, en este caso se va cambiando la tonalidad
      if k%100==0:
        color=(255-int((k/100)*25+5),255-int((k/100)*25+5),255)
      if abs(res)>paro:
        break
    out.putpixel((i,j),color)
plt.imshow(out)
```

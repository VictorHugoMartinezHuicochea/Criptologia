# Importamos Librerias


```python
import PIL
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt
```

# Ingresamos las funciones a visualizar

```python
f1=lambda x,y:(1/2*x,1/2*y)

def f2(x,y):
    return (1/2*x+1/2,1/2*y)

f3=lambda x,y:(1/2*x+1/4,1/2*y+1/2)

funciones=[f1,f2,f3]
```

# Funcion Auxiliar

```python
def mover_pixeles(x, y):
  mov=[]
  for i in range(len(funciones)):
    #print(i)
    res=funciones[i](x/alto,y/ancho)
    res=(int(res[0]*alto),int(res[1]*ancho))
    mov.append(res)
  return mov
```

# Realizacion del SFI

## Iteracion 1

```python
# Creamos la iteracion 1 del SFI
alto,ancho=500,500
out = Image.new("1", (alto, ancho),1)
#Dibujamos cuadrado
rect=ImageDraw.Draw(out)
rect.rectangle([(0,0),(499,499)], outline="black", width=1)
#Mostramos Iteracion 1
plt.imshow(out)
```

## Aplicacion del SFI

```python
#Guaardaremos cada iteracion para observar su evolucion
fig = plt.figure(figsize=(16, 16))
#Creamos imagen auxiliar
mod = Image.new("1", (alto, ancho),1)

#Iniciamos las iteraciones
for j in range(12):
  # Aplicar la función a cada píxel
  for y in range(alto):
    for x in range(ancho):
      #Verificamos que sea parte del conjunto compacto
      color = out.getpixel((x, y))
      if color == 0:
        #Movemos los pixeles
        mov=mover_pixeles(x, y)
        #Agregamos cada pixel al auxiliar
        for i in mov:
          #Verificamos que estan dentro de los limites de la imagen
          if i[0]<alto and i[1]<ancho:
            mod.putpixel(i, 0)
  #Guardamos el elemento nuevo
  out=mod.copy()
  fig.add_subplot(3, 4, j+1)
  plt.imshow(out)
  #Limpiamos el auxiliar
  mod = Image.new("1", (alto, ancho),1)
#Mostramos resultados
plt.imshow(out)
```
